### Data Preprocessing

The dataset was preprocessed using Linux `awk` commands to:
- Remove metadata lines
- Filter chromosome 21 variants
- Extract effect weights and harmonized positions

The cleaned TSV file was then loaded into pandas for analysis.

####Exploratory Data Analysis
Exploratory analysis was performed to understand the distribution and characteristics of effect weights for variants mapped to chromosome 21.

After preprocessing, the dataset contained only harmonized variants from chromosome 21 along with their corresponding effect weights and genomic positions. The primary focus of the exploratory analysis was the effect_weight column, as it represents the contribution of each genetic variant to the polygenic score.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the TSV file
df = pd.read_csv("chr21_effect_weights.tsv", sep="\t", header=None)

# Rename columns for clarity
df.columns = ["effect_weight", "hm_chr", "hm_pos"]

# Show first 5 rows
df.head()
```

```python
# Calculate summary statistics
mean_val = df["effect_weight"].mean()
median_val = df["effect_weight"].median()

# Create professional histogram
plt.figure(figsize=(8, 6))
plt.hist(df["effect_weight"], bins=25, edgecolor='black')

# Add mean and median lines
plt.axvline(mean_val, linewidth=2,
            label=f"Mean = {mean_val:.4f}")
plt.axvline(median_val, linewidth=2,
            label=f"Median = {median_val:.4f}")

# Labels and title
plt.xlabel("Effect Weight")
plt.ylabel("Number of Variants")
plt.title("Distribution of Effect Weights for Chromosome 21 Variants")

# Grid and legend
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()

plt.tight_layout()

# saving the image
plt.savefig("chr21_effect_weight_distribution.png",
            dpi=300, bbox_inches="tight")


plt.show()

# Summary statistics
df["effect_weight"].describe()
```
